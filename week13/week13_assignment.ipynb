### 2. Data Diversity

Use Python to identify how many unique values each column in your data has.  Use Python to identify which column has the greatest number of distinct values (aka the most diversity)?  If your data has a unique identifier (e.g. a person or visit or record ID) then find the column with the next most number of unique values.

```python
# Identify the number of unique values in each column
unique_counts = all_states.nunique()

# Display the number of unique values in each column
print("Number of unique values in each column:")
print(unique_counts)

# Find the column with the greatest number of distinct values
max_unique_column = unique_counts.idxmax()
max_unique_count = unique_counts.max()

print("\nColumn with the most unique values:")
print(f"Column: {max_unique_column}")
print(f"Number of unique values: {max_unique_count}")

# If there's a unique identifier column, find the column with the next most number of unique values
unique_counts.drop(max_unique_column, inplace=True)
second_max_unique_column = unique_counts.idxmax()
second_max_unique_count = unique_counts.max()

print("\nColumn with the second most unique values:")
print(f"Column: {second_max_unique_column}")
print(f"Number of unique values: {second_max_unique_count}")
```

### 4. Range of Values

Your data almost certainly has some numeric columns that can be summed or averaged.  Create a histogram that shows the distribution of values for this column.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Assuming 'DUI' and 'Population' are the column names in your DataFrame
# Replace them with the actual column names if they are different
columns_of_interest = ['DUI', 'Population']
# Create histograms for each specified column
for column in columns_of_interest:
    plt.figure(figsize=(8, 4))
    plt.hist(all_states[column], bins=15, color='blue', edgecolor='black')
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
dui_data = pd.read_csv('DUI.csv')

# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(dui_data[['DUI', 'Fatalities', 'Population']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of DUI, Fatalities, and Population')
plt.show()

# Boxplot of DUI incidents by Sunday Sales category
plt.figure(figsize=(10, 6))
sns.boxplot(x='Sunday.Sales', y='DUI', data=dui_data, palette='Set2')
plt.title('DUI Incidents by Sunday Sales Laws')
plt.show()

# Scatterplot: DUI vs. Fatalities
plt.figure(figsize=(8, 6))
sns.scatterplot(x='DUI', y='Fatalities', hue='Sunday.Sales', data=dui_data, palette='Set1')
plt.title('DUI vs. Fatalities, Colored by Sunday Sales Laws')
plt.xlabel('Number of DUIs')
plt.ylabel('Number of Fatalities')
plt.legend(title='Sunday Sales')
plt.show()
```
